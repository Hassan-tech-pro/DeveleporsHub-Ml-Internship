## 3. Scatter Plots: Feature Relationships
Scatter plots help us visualize relationships between pairs of features. We'll examine:
- Sepal length vs. sepal width
- Petal length vs. petal width

```python
# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 6))
plt.scatter(iris["sepal_length"], iris["sepal_width"], alpha=0.6, s=100, color='steelblue', edgecolors='black', linewidth=0.5)
plt.title("Iris Sepal: Length vs Width", fontsize=14, fontweight='bold')
plt.xlabel("Sepal Length (cm)")
plt.ylabel("Sepal Width (cm)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot: Petal Length vs Petal Width
plt.figure(figsize=(8, 6))
plt.scatter(iris["petal_length"], iris["petal_width"], alpha=0.6, s=100, color='coral', edgecolors='darkred', linewidth=0.5)
plt.title("Iris Petal: Length vs Width", fontsize=14, fontweight='bold')
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

## 4. Histograms: Value Distributions
Histograms reveal the frequency distribution of individual features:

```python
plt.figure(figsize=(8, 5))
plt.hist(iris.sepal_width, bins=15, color='skyblue', edgecolor='black', alpha=0.7)

plt.title('Distribution of Sepal Width', fontsize=14, fontweight='bold')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(iris.petal_length, bins=15, color='lightcoral', edgecolor='darkred', alpha=0.7)

plt.title('Distribution of Petal Length', fontsize=14, fontweight='bold')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.grid(axis='y', alpha=0.3)
plt.show()
```

## 5. Box Plots: Outlier Detection
Box plots provide a visual summary of data distribution and help identify potential outliers:

```python
# Check the columns to understand data structure
print("Dataset Columns:")
print(iris.columns.tolist())
print("\nDataset Shape:", iris.shape)
```

```python
plt.figure(figsize=(10, 6))
plt.boxplot(iris.drop("species", axis=1), 
            labels=['Sepal Length', 'Sepal Width', 'Petal Length', 'Petal Width'],
            patch_artist=True,
            boxprops=dict(facecolor='lightblue', alpha=0.7),
            medianprops=dict(color='red', linewidth=2))

plt.title('Iris Dataset - Box Plot Analysis (Outlier Detection)', fontsize=14, fontweight='bold')
plt.ylabel('Values (cm)')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
